```python
import numpy as np
```

```python
import torch
```

```python
import torch.nn as nn
import torch.optim as optim
from sklearn.preprocessing import StandardScaler
from torch.utils.data import DataLoader, TensorDataset
```

```python
features = df.drop(columns=["payloadguid"])
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)
X_train = torch.tensor(features_scaled, dtype=torch.float32)
batch_size = 32
payload_ids = df["payloadguid"]
train_loader = DataLoader(TensorDataset(X_train), batch_size=batch_size, shuffle=True)
class Autoencoder(nn.Module):
    def __init__(self, input_dim):
        super(Autoencoder, self).__init__()
        self.encoder = nn.Sequential(
            nn.Linear(input_dim, 16),
            nn.ReLU(),
            nn.Linear(16, 8),
            nn.ReLU(),
            nn.Linear(8, 4)
        )
        self.decoder = nn.Sequential(
            nn.Linear(4, 8),
            nn.ReLU(),
            nn.Linear(8, 16),
            nn.ReLU(),
            nn.Linear(16, input_dim)
        )

    def forward(self, x):
        encoded = self.encoder(x)
        decoded = self.decoder(encoded)
        return decoded
input_dim = X_train.shape[1]
model = Autoencoder(input_dim)
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
epochs = 100
for epoch in range(epochs):
    for data in train_loader:
        inputs = data[0]
        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, inputs)  # Reconstruction loss
        loss.backward()
        optimizer.step()
    if epoch % 10 == 0:
        print(f"Epoch [{epoch}/{epochs}], Loss: {loss.item():.4f}")

with torch.no_grad():
    reconstructions = model(X_train)
    reconstruction_errors = torch.mean((X_train - reconstructions) ** 2, dim=1)
threshold = reconstruction_errors.mean() + 2 * reconstruction_errors.std()
anomalies = reconstruction_errors > threshold
anomalous_samples = df[anomalies.numpy()]
anomalous_samples["payloadguid"] = payload_ids[anomalies.numpy()]
print("Detected Anomalous Samples:")
print(anomalous_samples[["payloadguid"] + list(features.columns)])
```
