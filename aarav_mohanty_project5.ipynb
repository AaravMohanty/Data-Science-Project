```python
# Importing pandas, matplotlib, and numpy
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# To display max columns
pd.set_option('display.max_columns', None)
```

```python
# Importing the death records and saving in data frame
myDeathRecordsDF = pd.read_csv("/anvil/projects/tdm/data/death_records/DeathRecords.csv")
```

```python
# Printing head
myDeathRecordsDF.head()
```

```python
# Creating a subset of only the women in these death records
womenDeathRecords = myDeathRecordsDF[myDeathRecordsDF['Sex'] == 'F']
```

```python
# Printing the women death records
womenDeathRecords.head()
```

```python
# Creating a value count table for the ages of the women in the death records
womenAgeFrequency = womenDeathRecords['Age'].value_counts()
```

```python
# Sorting the value_counts()
womenAgeFrequency = womenAgeFrequency.sort_index()
womenAgeFrequency
```

```python
# Plotting the age frequency
plt.bar(womenAgeFrequency.index, womenAgeFrequency)
```

```python
# Creating a subset of only the women in these death records that are valid
womenDeathRecordsKnownAge = myDeathRecordsDF[(myDeathRecordsDF['Sex'] == 'F') & (myDeathRecordsDF['Age'] != 999)]
```

```python
# Printing the women death records
womenDeathRecordsKnownAge.head()
```

```python
# Creating a value count table for the ages of the women in the death records 
womenAgeFrequencyKnown = womenDeathRecordsKnownAge['Age'].value_counts()
```

```python
# Sorting the value_counts()
womenAgeFrequencyKnown = womenAgeFrequencyKnown.sort_index()
womenAgeFrequencyKnown
```

```python
# Plotting the value counts 
plt.bar(womenAgeFrequencyKnown.index, womenAgeFrequencyKnown)
```
